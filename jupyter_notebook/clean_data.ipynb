```python
import pandas as pd
import numpy as np

df = pd.read_csv('../dataset/diabetes_early_stage.csv')

df.info()
```

```python
print("Missing values:\n", df.isnull().sum())

duplicates = df.duplicated().sum()
print(f"\nFound {duplicates} duplicate rows. Dropping them...")
df = df.drop_duplicates()
```

```python
# Gender: Male -> 1, Female -> 0
df['Gender'] = np.where(df['Gender'] == 'Male', 1, 0)

# class: Positive -> 1, Negative -> 0
df['class'] = np.where(df['class'] == 'Positive', 1, 0)

# All other columns with 'Yes' / 'No'
for col in df.columns:
    if df[col].dtype == 'object':
        # Convert 'Yes' to 1, everything else (e.g., 'No') to 0
        df[col] = np.where(df[col] == 'Yes', 1, 0)

df.head()
```

```python
df.to_csv('../dataset/diabetes_early_stage_cleaned.csv', index=False)
print("Cleaned dataset saved as 'diabetes_early_stage_cleaned.csv'")
```
